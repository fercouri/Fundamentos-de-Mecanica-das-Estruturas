```python
!pip install anastruct
from anastruct import SystemElements

ss = SystemElements()

# Elementos
ss.add_element(location=[[0, 0], [2.69, 2]]) # AD
ss.add_element(location=[[2.69, 2], [4, 3]], spring={2:0}) # DC
ss.add_element(location=[[4, 3], [5.31, 2]]) # CE
ss.add_element(location=[[5.31, 2], [8, 0]]) # EB
ss.add_element(location=[[2.69, 2], [5.31, 2]]) # Tirante DE

# Apoios
ss.add_support_hinged(node_id=1) # Apoio em A
ss.add_support_hinged(node_id=5)  # Apoio em B

# Carga distribuda na barra horizontal (DE)
ss.q_load(element_id=1, q=-0.96, direction='y')
ss.q_load(element_id=2, q=-0.96, direction='y')
ss.q_load(element_id=3, q=-0.96, direction='y')
ss.q_load(element_id=4, q=-0.96, direction='y')

# Resolver
ss.solve()

# Plots
ss.show_structure()
ss.show_reaction_force()
ss.show_axial_force()
ss.show_shear_force()
ss.show_bending_moment()
```
